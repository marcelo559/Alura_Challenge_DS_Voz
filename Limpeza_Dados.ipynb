# Organizando o DataFrame

```python
# Função para realizar a normalização dos objetos json e a junção das informações em um DataFrame

def normalize_json(dataframe):
    return_dataframe = pd.DataFrame()
    for column in list(dados.columns[2:]):
        dataframe_column = pd.json_normalize(dataframe[column])
        return_dataframe = pd.concat([return_dataframe, dataframe_column], axis=1)
    
    return pd.concat([dataframe[list(dados.columns[:2])], return_dataframe], axis=1)
```

```python
dados = normalize_json(dados)
dados
```

# Avaliando e corrigindo os dados

```python
dados.info()
```

```python
# O método value_counts() verifica nas colunas possíveis categorias com nomes incorretos ou incoerentes

for coluna in dados.columns:
    print('---')
    print(dados[coluna].value_counts())
```

```python
# Conseguimos verificar que a coluna CHUR tem dados sem classificação, o que é um problema e precisa ser removido

dados['Churn'].value_counts()
```

```python
# Removendo valores vazios em Churn

dados = dados[dados['Churn']!= '']
dados.reset_index(drop=True, inplace=True)
```

```python
# Verificando ser a ariável Churn não apresenta mais a classe com nome vazio

dados['Churn'].value_counts()
```

```python
# Vamos selecionar todas as linhas onde o tenure = 0, ou seja, os clientes que assinaram o serviço por 0 meses
# e mostrar os resultados para as colunas Charges.Total e Charges.Monthly.

dados.query('tenure == 0')[['Charges.Total', 'Charges.Monthly', 'tenure']]
```

```python
# Verificamos que quando tenure = 0, os dados de Charges.Total é vazio (' ').
# Vamos agora selecionar os dados onde Charges.Total = ' ', mostrando os resultados para Charges.Monthly e tenure.
# Pegando todas as linhas onde a coluna "Charges.Total" é vazia.

dados[dados['Charges.Total'] == ' '][['Charges.Total', 'Charges.Monthly', 'tenure']]
```

```python
# Percebe-se que todas as linhas de Charges.Total que estão vazias é porque o cliente não assinou nem por um mês. Temos que preencher esse valor com o mesmo valor que está presente em "Charges.Monthly" pois esse representa o total.

idx = dados[dados['Charges.Total'] == ' '].index
dados.loc[idx, "Charges.Total"] = dados.loc[idx, "Charges.Monthly"]
```

```python
dados.query('tenure == 0')[['Charges.Total', 'Charges.Monthly', 'tenure']]
```

```python
# Vamos modificar o tipo da variável para float, uma vez que o tipo estava como object.

dados['Charges.Total'] = dados['Charges.Total'].astype('float64')
```

# Extra
## Traduzindo colunas do DataFrame

```python
# Listagem das colunas atuais para fazer a tradução das colunas, pensando em pessoas que não tem intimidade com o inglês
dados.info()
```

```python
# Criando um novo DataFrame com o De : Para das novas colunas

dados_traduzidos = dados.rename(columns={
    'gender':'Genero',
    'SeniorCitizen':'MelhorIdade',
    'Partner':'Conjuge',
    'Dependents':'Dependentes',
    'tenure':'MesesContrato',
    'PhoneService':'ServicoTelefonico',
    'MultipleLines':'MultiplosTelefones',
    'InternetService':'ServicoInternet',
    'OnlineSecurity':'SergurancaOnline',
    'OnlineBackup':'BackupOnline',
    'DeviceProtection':'ProtecaoDevice',
    'Contract':'TipoContrato',
    'PaperlessBilling':'TipoDaFatura',
    'PaymentMethod':'FormaDePagamento',
    'Charges.Monthly':'Cobranca.Mes',
    'Charges.Total':'Cobranca.Total'
})
```

```python
# Conferir se o RENAME está correto e saiu tudo conforme esperado

dados_traduzidos.info()
```

## Criar uma nova coluna com o valor diário para possíveis insights

```python
dados_traduzidos[dados_traduzidos['MesesContrato'] == 0]
```

```python
# Crie uma nova coluna contendo os gastos diários do(a) cliente
# Nesta nova coluna, na primeira tentativa, vou considerar que todos os meses tem 30 dias

for index, row in dados_traduzidos.iterrows():
    dados_traduzidos.loc[index, 'GastoDiario30Dias'] = row['Cobranca.Mes'] / 30

dados_traduzidos
```

## Criar a mesma coluna com valores diário com pensando de uma outra forma

```python
# Utilizando a mesma ideia de criar uma nova coluna contendo os gastos diários
# Acredito te uma outra forma de conseguir esse resultado, talvez um pouco mais acetiva quando falamos em gastos diários (365 dias no ano)
# Considerando que nem todos os meses tem 30 dias, e considerando que temos 365 dias (desconsiderando os anos bisextos)
# Pensei em (valor mensal * 12 meses) / 365 dias do ano 

for index, row in dados_traduzidos.iterrows():
  dados_traduzidos.loc[index, 'GastoDiario365Dias'] = (row['Cobranca.Mes'] * 12) / 365

dados_traduzidos
```
